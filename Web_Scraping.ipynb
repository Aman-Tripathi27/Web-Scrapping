```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get("https://www.ambitionbox.com/list-of-companies?page=1", headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_="companyCardWrapper")
```

```python
len(company)
```

```python
name = []
rating = []
reviews = []
cinfo = []
for i in company:

    name.append(i.find('h2').text.strip())
    rating.append(i.find('div', class_= "companyCardWrapper__companyRatingWrapper").text.strip())
    reviews.append(i.find('span', class_="companyCardWrapper__ActionCount").text.strip())
    cinfo.append(i.find('span', class_="companyCardWrapper__interLinking").text.strip())


d = {'name':name,
     'rating':rating,
     'reviews':reviews,
     'cinfo':cinfo}

df = pd.DataFrame(d)
df
```

```python
final=pd.DataFrame()

for j in range(1,500):


    url="https://www.ambitionbox.com/list-of-companies?page={}".format(j)

    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text


    soup = BeautifulSoup(webpage, 'lxml')

    company = soup.find_all('div', class_="companyCardWrapper")

    name = []
    rating = []
    reviews = []
    cinfo = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_= "companyCardWrapper__companyRatingWrapper").text.strip())
        reviews.append(i.find('span', class_="companyCardWrapper__ActionCount").text.strip())
        cinfo.append(i.find('span', class_="companyCardWrapper__interLinking").text.strip())


    d = {'name':name, 'rating':rating, 'reviews':reviews, 'cinfo':cinfo}
    df = pd.DataFrame(d)


    final =final.append(df,ignore_index=True)




```
